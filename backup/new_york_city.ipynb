# New York City Neighborhood mining

This notebook combines data sources (wikipedia and geocoder) to assemble a list of neighborhoods and map them to their coordinates for futher processing.

```python
import pandas as pd
import requests

import json
from pandas.io.json import json_normalize

from geocoder import enrich_neighborhoods_with_geocoder, map_neighborhoods
```

### Download geo.nyu.edu neighborhoods dataset

```python
# results = pd.read_json()
# results

r  = requests.get('https://cocl.us/new_york_dataset')
data = r.text
results = json.loads(data)
print(results)
```

```python
df = pd.DataFrame(columns=['Neighborhood', 'Latitude', 'Longitude'])

neighborhoods = results['features']

for data in neighborhoods:
    neighborhood_name = data['properties']['name']
        
    neighborhood_latlon = data['geometry']['coordinates']
    neighborhood_lat = neighborhood_latlon[1]
    neighborhood_lon = neighborhood_latlon[0]
    
    df = df.append({'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)

df
```
