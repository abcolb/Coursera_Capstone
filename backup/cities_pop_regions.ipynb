```python
import pandas as pd

import json
import requests
```

```python
url = 'https://public.opendatasoft.com/api/records/1.0/search/?dataset=worldcitiespop&rows=10000&sort=population&facet=country'
```

```python
data = json.loads(requests.get(url).text)
data.keys()
```

```python
parsed = ({
    'city': r.get('fields').get('city'), 
    'region':  r.get('fields').get('region'), 
    'country':  r.get('fields').get('country'),
    'population': r.get('fields').get('population'),
    'geometry':  r.get('geometry'),
} for r in data.get('records'))

cities_pop_regions = pd.DataFrame(parsed)
cities_pop_regions
```

```python
cities_pop_regions[cities_pop_regions.population > 500000]
```

```python
cities_pop_regions[cities_pop_regions.population > 500000].to_csv('data/cities_pop_regions.csv')
```
